```python
import os

excluded = ["unity_game/.", "unity_game\\Libs", "unity_game\\CommandsSystem\\Generated"]
def is_excluded(path):
    for i in excluded:
        if path.startswith(i):
            return True
    return False
    
cs = []
    
for root, subdirs, files in os.walk("unity_game"):
    if is_excluded(root): continue
    for file in files:
        if not file.endswith(".c") or not file.endswith(".h"): continue
        cs.append(root + "/" + file)
cs
```

```python
classes = {} # class name -> (classtype, name, summary, [fields], [properties], [methods])


def clear_empty(s):
    return [x for x in s if x != '' and x != " "]


def handle_file(file):
    with open(file, encoding="utf-8") as f:
        namespace = ""
        classname = ""
        lastclass = ""
        lastclasstype = ""
        classtype = ""
        curTokenType = "none"
        token = ""
        summary = ""
        class_brackets_balance = 0
        brackets_balance = 0

        for line in f.read().split("\n"):
            if 'namespace' in line:
                namespace = (" " + line).split("namespace")[1].split("{")[0].strip()
                continue
            brackets_balance += line.count("{") - line.count("}")
            if brackets_balance < class_brackets_balance:
                classname = lastclass
                classtype = lastclasstype
                class_brackets_balance -= 1
            if 'class ' in line or 'struct ' in line or 'interface ' in line or 'enum ' in line:
                class_brackets_balance = brackets_balance
                lastclass = classname
                lastclasstype = classtype
                classtype = ""
                if 'class' in line: classtype = 'class'
                if 'struct' in line: classtype = 'struct'
                if 'interface' in line: classtype = 'interface'
                if 'enum' in line: classtype = 'enum'
                
                classname = line.split(classtype)[1].split()[0]
                if lastclass != "":
                    classname = lastclass + "." + classname
                if namespace != "":
                    classname = namespace + "." + classname
                classes[classname] = (classtype, classname, summary, [], [], [])
                summary = ""
                continue

            if "///" in line:
                if "<summary>" in line:
                    curTokenType = summary
                    token = ""
                elif "</summary>" in line:
                    curTokenType = "none"
                    summary = token.strip()
                else:
                    token += line.split("///")[1].strip() + " "
                continue

            if "//" in line or "delegate" in line:
                continue
        
            if classname == '' or classtype == "enum": continue
            
            line = line.replace("static", "").replace("override", "").replace(";", "").replace("virtual", "").replace("abstract", "").replace("new", "").replace("readonly", "").replace("async", "")
#             print(brackets_balance, classname, line)
            
            # method
            if brackets_balance - line.count("{") + line.count("}") == 1 and "(" in line and not '=' in line:
                curTokenType = "method"
                mod = "none"
                if "public" in line: mod = "public"
                if "private" in line: mod = "private"
                if "internal" in line: mod = "internal"
                if "protected" in line: mod = "protected"
                line2 = line
                parts = []
                if mod == "none":
                    mod = "private"
                    parts = clear_empty(line2.split("(")[0].split(" "))
                else:
                    parts = clear_empty(line2.split("(")[0].split(" "))[1:]
                if len(parts) < 2: # means it is constructor
                    retval = "конструктор"
                elif len(parts) == 2:
                    retval = parts[0]
                else:
                    raise Exception("strange amount of parts " + str(parts) + "  " + line)
                name = parts[-1].split("(")[0]
                args = []
                
            # field
            elif brackets_balance == 1 and line.strip() != "":
                parts = clear_empty(line.split())
                if len(parts) >= 3:
                    mod = "none"
                    if "public" in line: mod = "public"
                    if "private" in line: mod = "private"
                    if "internal" in line: mod = "internal"
                    if "protected" in line: mod = "protected"
                    if mod == "none":
                        mod = "private"
                    else:
                        parts = parts[1:]
                    typ = parts[0]
                    name = parts[1]
                    if "{" in line:
                        curTokenType = "property"
                        get_mod = set_mod = mod
                    else:
                        classes[classname][3].append((summary, mod, typ, name))
            
            if curTokenType == "property" and 'get' in line:
                if 'private' in line.split('get')[0]:
                    get_mod = 'private'
            if curTokenType == "property" and 'set' in line:
                if 'private set' in line:
                    set_mod = 'private'
                    
            if curTokenType == "property" and "}" in line:
                curTokenType = "none"
                classes[classname][4].append((summary, (get_mod + " get; " + set_mod + " set").strip(), typ, name))
                print(classes[classname][4][-1])
                        
            if curTokenType == "method":
                line2 = line
                if '(' in line2: line2 = line.split("(")[1]
                if ')' in line2: line2 = line2.split(")")[0]
                line2 = line2.split(",")
                line2 = [x.strip() for x in line2 if x.strip() != '']
                args += line2
            if curTokenType == "method" and ")" in line:
                classes[classname][5].append((summary, mod, retval, name, args))
                summary = ""
                mod = ""
                retval = ""
                name = ""
                args = []
                curTokenType = "none"
            
for file in cs:
    print(file)
    handle_file(file)
classes
```

```python
# class name -> (classtype, name, summary, [fields], [properties], [methods])
# method (summary, mod, retval, name, args)
# field (summary, mod, typ, name)
# property (summary, (get_mod + " get; " + set_mod + " set").strip(), typ, name)



from odf.opendocument import OpenDocumentText
from odf.style import Style, TextProperties, ParagraphProperties
from odf.style import TableColumnProperties, TableRowProperties, TableCellProperties
from odf.text import P
from odf.table import Table, TableColumn, TableRow, TableCell

doc = OpenDocumentText()

centertext = Style(name="Center text", family="paragraph")
centertext.addElement(ParagraphProperties(textalign="center"))
centertext.addElement(TextProperties(attributes={"fontfamily":"Times New Roman"}))
doc.styles.addElement(centertext)


righttext = Style(name="Right Text", family="paragraph")
righttext.addElement(ParagraphProperties(textalign="right"))
righttext.addElement(TextProperties(attributes={"fontfamily":"Times New Roman"}))
doc.styles.addElement(righttext)

tablecontents = Style(name="Table Contents", family="paragraph")
tablecontents.addElement(ParagraphProperties(numberlines="false", linenumber="0"))
tablecontents.addElement(ParagraphProperties(textalign="left"))
tablecontents.addElement(TextProperties(attributes={"fontfamily":"Times New Roman"}))
tablecontents.addElement(TableCellProperties(attributes={"border":"true"}))

doc.styles.addElement(tablecontents)

centeredtablecontents = Style(name="Centered Table Contents", family="paragraph")
centeredtablecontents.addElement(ParagraphProperties(numberlines="false", linenumber="0"))
centeredtablecontents.addElement(ParagraphProperties(textalign="center"))
centeredtablecontents.addElement(TextProperties(attributes={"fontfamily":"Times New Roman"}))
doc.styles.addElement(centeredtablecontents)

# Create automatic styles for the column widths.
# We want two different widths, one in inches, the other one in metric.
# ODF Standard section 15.9.1
widthshort = Style(name="Wshort", family="table-column")
widthshort.addElement(TableColumnProperties(columnwidth="3cm"))
doc.automaticstyles.addElement(widthshort)

widthshort2 = Style(name="Wshort2", family="table-column")
widthshort2.addElement(TableColumnProperties(columnwidth="2cm"))
doc.automaticstyles.addElement(widthshort2)


widthshort25 = Style(name="Wshort25", family="table-column")
widthshort25.addElement(TableColumnProperties(columnwidth="2.3cm"))
doc.automaticstyles.addElement(widthshort25)


widthshort4 = Style(name="Wshort4", family="table-column")
widthshort4.addElement(TableColumnProperties(columnwidth="4cm"))
doc.automaticstyles.addElement(widthshort4)

def add_text_to_row(row, text, cnt=1, style=tablecontents):
    tc = TableCell(numbercolumnsspanned=cnt)
    tr.addElement(tc)
    p = P(stylename=style,text=text)
    tc.addElement(p)

classtype_as_str1 = {
    "class": "класса",
    "interface": "интерфейса",
    "struct": "структуры",
    "enum": "перечисления",
}


doc.text.addElement(P(stylename=righttext, text="Таблица"))
doc.text.addElement(P(stylename=centertext, text="Описание и функцинальное назначение классов"))

    
table = Table()
table.addElement(TableColumn(stylename=widthshort2))
table.addElement(TableColumn(stylename=widthshort4))
tr = TableRow()
add_text_to_row(tr, "Классы", 1, style=centeredtablecontents)
add_text_to_row(tr, "Назначение", 1, style=centeredtablecontents)
table.addElement(tr)
for cl in sorted(classes.values(), key=lambda x: x[1]):
    tr = TableRow()
    #tr.addElement(TableRowProperties(keeptogether=True))
    add_text_to_row(tr, cl[1])
    add_text_to_row(tr, cl[2])
    table.addElement(tr)

doc.text.addElement(table)
                    
    
table_num = 1

for cl in sorted(classes.values(), key=lambda x: x[1]):
    if len(cl[3]) == 0 and len(cl[4]) == 0 and len(cl[5]) == 0: continue
    doc.text.addElement(P())
    doc.text.addElement(P(stylename=righttext, text="Таблица 3." + str(table_num)))
    doc.text.addElement(P(stylename=centertext, text="Описание полей, методов и свойств " + classtype_as_str1[cl[0]] + " " + cl[1]))                    
    table = Table()
    table.addElement(TableColumn(stylename=widthshort))
    table.addElement(TableColumn(stylename=widthshort25))
    table.addElement(TableColumn(stylename=widthshort2))
    table.addElement(TableColumn(stylename=widthshort))
    table.addElement(TableColumn(stylename=widthshort4))
    
    
    if len(cl[3]) > 0:
        tr = TableRow()
        add_text_to_row(tr, "Поля  "+ classtype_as_str1[cl[0]] + " " + cl[1], 5, style=centeredtablecontents)
        table.addElement(tr)

        tr = TableRow()
        add_text_to_row(tr, "Имя", 1)
        add_text_to_row(tr, "Модификатор доступа", 1)
        add_text_to_row(tr, "Тип", 1)
        add_text_to_row(tr, "Назначение", 2)
        table.addElement(tr)

        for field in cl[3]:
            tr = TableRow()
            add_text_to_row(tr, field[3])
            add_text_to_row(tr, field[1])
            add_text_to_row(tr, field[2])
            add_text_to_row(tr, field[0], 2)
            table.addElement(tr)

    
    
    
    
    if len(cl[4]) > 0:
        tr = TableRow()
        add_text_to_row(tr, "Свойства  "+ classtype_as_str1[cl[0]] + " " + cl[1], 5, style=centeredtablecontents)
        table.addElement(tr)

        tr = TableRow()
        add_text_to_row(tr, "Имя", 1)
        add_text_to_row(tr, "Модификатор доступа", 1)
        add_text_to_row(tr, "Тип", 1)
        add_text_to_row(tr, "Назначение", 2)
        table.addElement(tr)

        for prop in cl[4]:
            tr = TableRow()
            add_text_to_row(tr, prop[3])
            add_text_to_row(tr, prop[1])
            add_text_to_row(tr, prop[2])
            add_text_to_row(tr, prop[0], 2)
            table.addElement(tr)
    
    
    
    
    
    if len(cl[5]) > 0:
        tr = TableRow()
        add_text_to_row(tr, "Методы  "+ classtype_as_str1[cl[0]] + " " + cl[1], 5, style=centeredtablecontents)
        table.addElement(tr)

        tr = TableRow()
        add_text_to_row(tr, "Имя", 1)
        add_text_to_row(tr, "Модификатор доступа", 1)
        add_text_to_row(tr, "Тип", 1)
        add_text_to_row(tr, "Аргументы", 1)
        add_text_to_row(tr, "Назначение", 1)
        table.addElement(tr)

        for method in cl[5]:
            tr = TableRow()
            add_text_to_row(tr, method[3])
            add_text_to_row(tr, method[1])
            add_text_to_row(tr, method[2])
            add_text_to_row(tr, ", ".join(method[4]))
            add_text_to_row(tr, method[0])
            table.addElement(tr)
    
    
    doc.text.addElement(table)
    table_num += 1



# def create_tables_for_class(cl):
#     table = Table()
#     table.addElement(TableColumn(numbercolumnsrepeated=4,stylename=widthshort))
#     table.addElement(TableColumn(numbercolumnsrepeated=3,stylename=widthshort))

#     for line in cl.met:
#         rec = line.strip().split(":")
#         tr = TableRow()
#         table.addElement(tr)
#         for val in rec:
#             tc = TableCell()
#             tr.addElement(tc)
#             p = P(stylename=tablecontents,text=val)
#             tc.addElement(p)

doc.save("Пояснительная Записка", True)
```
